##Detect a Vehicle


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


image_path = '/content/Truck_template.jpg'
main_image_path = '/content/Find_Truck_color.jpg'


template = cv2.imread(image_path, 0)
main_image = cv2.imread(main_image_path, 0)


if template is None:
    raise FileNotFoundError(f"Error: {image_path} not found or failed to load.")
if main_image is None:
    raise FileNotFoundError(f"Error: {main_image_path} not found or failed to load.")

# Perform template matching
result = cv2.matchTemplate(main_image, template, cv2.TM_CCOEFF_NORMED)

# Get the location of the best match
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
top_left = max_loc
bottom_right = (top_left[0] + template.shape[1], top_left[1] + template.shape[0])

# Draw a rectangle around the matched region
cv2.rectangle(main_image, top_left, bottom_right, 255, 2)

# Display the images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(template, cmap='gray')
plt.title('Template Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(main_image, cmap='gray')
plt.title('Detected Template in Main Image')
plt.axis('off')

plt.show()
```
